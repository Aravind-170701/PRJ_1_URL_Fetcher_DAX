# Google Map Scraping using Selenium 

```python
#Importing Libraries

import numpy as np
import pandas as pd
from lxml import html, etree
from io import StringIO
import re
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import traceback
from selenium.webdriver.common.by import By
```

```python
# Filtering Dataframe containing State and City Information but no website URL

dataframe = pd.read_csv("input_data.csv",usecols=["infogroup_id","name","phone","street","city","labels.state","website","labels.location_parent_relationship"])

dataframe = dataframe[dataframe['website'].isna()]

filtered_dataframe = dataframe[(~dataframe['labels.state'].isna()) & (~dataframe['city'].isna())]

filtered_dataframe.reset_index(inplace = True, drop = True)
```

```python
#Map Scraper Class which scrapes google maps to fetch website URL based on the given query

class MapScraper:
    
    def __init__(self):
        
        options = Options()
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')
        self.driver = webdriver.Chrome(ChromeDriverManager().install(),options=options)

    def map_scrape(self, query: str) -> str:

        url = "https://www.google.com/maps/search/"+query
        self.driver.get(url)

        try:
            website = self.driver.find_element(By.CSS_SELECTOR, "[data-item-id='authority']").text
        except:
            website = 'invalid'

        self.driver.quit
        return website
    
```

```python
# Google Maps scraping based on a query (Name + Street Address + State)

scraper = MapScraper()

for i,row in filtered_dataframe.iterrows():

    query = str(row['name']) + " " + str(row['street']) + " " +str(row['labels.state'])
    website_link = scraper.map_scrape(query)
    print(website_link)

    filtered_dataframe.loc[i, 'website'] = website_link


filtered_dataframe
```
